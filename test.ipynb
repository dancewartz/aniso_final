```julia
include("HFModel.jl")
using .HFModel
using Plots
using ColorSchemes
using Colors
using LaTeXStrings
```

```julia
# Define variables explicitly
Df = 1.0      # Diffusion coefficient for frequencies
Dh = 1.0      # Diffusion coefficient for heights
v0 = 0.15      # Baseline speed
α = 0.25    # Parameter alpha
γ = 0.0      # Parameter gamma
λ = 20.0      # Parameter lambda
s0 = 0.25      # Source term for frequencies
f0 = 0.1      # Equilibrium frequency
β = 5.0      # Advection coefficient
L = 400.0      # Domain length
dx = 0.2      # Spatial step size
T = 80.0       # Total simulation time
dt = 0.01     # Time step size
thin_by = 100  # Thinning parameter

# Instantiate Parameters struct with the variables
par = Parameters(Df, Dh, v0, α, γ, λ, s0, f0, β, L, dx, T, dt, thin_by, :pulled, 0.0, 0.0)

frequencies = similar(par.xs)
frequencies .= 0;
frequencies[abs.(par.xs) .< 1] .= 1; 

heights = similar(par.xs)
heights .= 0;

@time result, record_times = run_simulation!(frequencies, heights, par);

println((2 * sqrt(s0 * Df))^2/2/λ)

# Define your colors
light_sky_blue = RGB(0/255, 174/255, 239/255)
yellow = RGB(249/255, 237/255, 50/255)

# Create a custom color gradient
custom_gradient = cgrad([yellow, light_sky_blue])

# Generate the plot
p = plot(par.xs, result[:,1:1:end,2], line_z=result[:,1:1:end,1], color=custom_gradient, label=:none,
         linewidth=4, xaxis=false, yaxis=false, colorbar=false, grid=false, title=L"$\alpha=$ %$(α)",
         titlefontsize=20)

savefig(p, "anisotropy/escaping.png")
savefig(p, "anisotropy/escaping.pdf")
savefig(p, "anisotropy/escaping.svg")
```

```julia
using Plots

# Define variables explicitly
Df = 1.0      # Diffusion coefficient for frequencies
Dh = 1.0      # Diffusion coefficient for heights
v0 = 1.0      # Baseline speed
γ = 0.0      # Parameter gamma
λ = 10.0      # Parameter lambda
s0 = 2.5      # Source term for frequencies
f0 = 0.1      # Equilibrium frequency
β = 10.0      # Advection coefficient
L = 2000.0      # Domain length
dx = 0.5      # Spatial step size
T = 550.0       # Total simulation time
dt = 0.01     # Time step size
thin_by = 1000  # Thinning parameter

println(Df * dt / dx^2)
```

```julia

# Define the range and step for alpha
α_values = 0:0.01:0.4
front_speeds = zeros(length(α_values))

# Loop over the range of alpha values
for (i, α) in enumerate(α_values)

    println(i)
    # Update alpha in the parameters
    par = Parameters(Df, Dh, v0, α, γ, λ, s0, f0, β, L, dx, T, dt, thin_by)

    # Initialize frequencies and heights
    frequencies = similar(par.xs)
    frequencies .= 0
    frequencies[par.xs .< -950] .= 1

    heights = similar(par.xs)
    heights .= 0

    # Run the simulation
    @time result, record_times = run_simulation!(frequencies, heights, par)

    # Calculate front locations
    front_locations = [sum(result[:, time_index, 1]) * par.dx for time_index in 1:length(record_times)]

    # Linear regression to find the invasion speed
    reg =  record_times .> par.T / 2
    X = hcat(record_times[reg], ones(length(record_times[reg])))
    invasion_speed = X \ front_locations[reg]

    # Store the first component of the invasion_speed
    front_speeds[i] = invasion_speed[1]
end

# Plot front speed as a function of alpha
plot(α_values, front_speeds, label="Front Speed", xlabel="Alpha (α)", ylabel="Front Speed", linewidth=2)
```

```julia
using Plots
using LaTeXStrings # For LaTeX-styled plot labels

# Define variables explicitly
Df = 1.0      # Diffusion coefficient for frequencies
Dh = 1.0      # Diffusion coefficient for heights
v0 = 1.0      # Baseline speed
λ = 10.0      # Parameter lambda
s0 = 4.0      # Source term for frequencies
f0 = 0.1      # Equilibrium frequency
β = 10.0      # Advection coefficient
L = 6000.0      # Domain length
dx = 0.5      # Spatial step size
T = 550.0       # Total simulation time
dt = 0.01     # Time step size
thin_by = 1000  # Thinning parameter

println(Df * dt / dx^2)
# Define the range and step for alpha and gamma
α_values = -0.4:0.04:0.4
γ_values = -0.4:0.1:0.4 # Example gamma values
front_speeds_matrix = zeros(length(α_values), length(γ_values)) # Store speeds for each α and γ

p = plot() # Initialize plot

for (j, γ) in enumerate(γ_values)
    front_speeds = zeros(length(α_values)) # Reset front speeds for new γ
    
    for (i, α) in enumerate(α_values)

        println(i, j)
        
        # Update alpha and gamma in the parameters
        par = Parameters(Df, Dh, v0, α, γ, λ, s0, f0, β, L, dx, T, dt, thin_by)
        # Initialize frequencies and heights
        frequencies = similar(par.xs)
        frequencies .= 0
        frequencies[par.xs .< 0] .= 1

        heights = similar(par.xs)
        heights .= 0

        # Run the simulation
        @time result, record_times = run_simulation!(frequencies, heights, par)

        # Calculate front locations
        front_locations = [sum(result[:, time_index, 1]) * par.dx for time_index in 1:length(record_times)]

        # Linear regression to find the invasion speed
        reg =  record_times .> par.T / 2
        X = hcat(record_times[reg], ones(length(record_times[reg])))
        invasion_speed = X \ front_locations[reg]
        
        # Store the calculated speed
        front_speeds[i] = invasion_speed[1]
    end

    front_speeds_matrix[:, j] = front_speeds # Store speeds for this gamma
    
    # Plotting for current γ value
    scatter!(p, α_values, front_speeds, label="γ=$γ", xlabel=L"Alpha (\alpha)", ylabel="Front Speed", linewidth=2)
end
```
